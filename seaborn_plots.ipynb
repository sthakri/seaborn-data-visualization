```python
# Import Required Libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Set theme
sns.set_theme(style="darkgrid")
```

```python
# Check available datasets in Seaborn
available_datasets = sns.get_dataset_names()
print(f"Available datasets: {available_datasets}")
```

```python
# ========== DATASET 1: TIPS DATASET ==========
# Load the tips dataset
tips = sns.load_dataset("tips")
print("✓ Tips dataset loaded successfully")
```

```python
# View full dataset
print("Full Tips Dataset shape:", tips.shape)
print("\n", tips)
```

```python
# First few rows
tips.head()
```

```python
# Data info
tips.info()
```

```python
# --- LINE PLOT ---
# Visualization: Relationship between total bill and tip
plt.figure(figsize=(10, 6))
sns.lineplot(x="total_bill", y="tip", data=tips, marker='o')
plt.title("Line Plot: Total Bill vs Tip", fontsize=12, fontweight='bold')
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.tight_layout()
plt.show()
```

```python
# --- SCATTER PLOT ---
# Shows individual data points
plt.figure(figsize=(10, 6))
sns.scatterplot(x="total_bill", y="tip", data=tips, s=100, alpha=0.6)
plt.title("Scatter Plot: Total Bill vs Tip", fontsize=12, fontweight='bold')
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.tight_layout()
plt.show()
```

```python
# --- BAR PLOT ---
# Better for viewing relationship between categorical and numerical data
plt.figure(figsize=(10, 6))
sns.barplot(x="day", y="total_bill", data=tips)
plt.title("Bar Plot: Average Total Bill by Day", fontsize=12, fontweight='bold')
plt.xlabel("Day of Week")
plt.ylabel("Total Bill ($)")
plt.tight_layout()
plt.show()
```

```python
# --- BOX PLOT ---
# Great for showing distribution and outliers
plt.figure(figsize=(10, 6))
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("Box Plot: Distribution of Total Bill by Day", fontsize=12, fontweight='bold')
plt.xlabel("Day of Week")
plt.ylabel("Total Bill ($)")
plt.tight_layout()
plt.show()
```

```python
# ========== DATASET 2: FLIGHTS DATASET ==========
# Load the flights dataset
flights = sns.load_dataset("flights")
print("✓ Flights dataset loaded successfully")
print(f"Shape: {flights.shape}")
```

```python
# View flights data
flights.head(10)
```

```python
# Pivot the data for heatmap (months as rows, years as columns)
pivot_data = flights.pivot(index="month", columns="year", values="passengers")
```

```python
# View pivoted data
pivot_data
```

```python
# --- HEATMAP ---
# Perfect for showing patterns across two categorical variables
plt.figure(figsize=(14, 8))
sns.heatmap(pivot_data, annot=True, fmt="d", cmap="YlGnBu", cbar_kws={"label": "Number of Passengers"})
plt.title('Heatmap: Airline Passengers by Month and Year (1949-1960)', fontsize=12, fontweight='bold')
plt.xlabel("Year")
plt.ylabel("Month")
plt.tight_layout()
plt.show()
```

```python
# ========== DATASET 3: CUSTOM DATAFRAME ==========
# Create a custom dataset
df = pd.DataFrame({
    "age": [22, 25, 47, 52, 46, 56, 55, 60, 34, 43],
    "salary": [25000, 27000, 52000, 60000, 58000, 62000, 61000, 65000, 38000, 45000],
    "gender": ["M", "F", "M", "F", "F", "M", "M", "F", "F", "M"]
})

print("✓ Custom DataFrame created successfully")
print(f"Shape: {df.shape}")
```

```python
# View the dataframe
df
```

```python
# --- SCATTER PLOT WITH HUE ---
# Add color dimension with 'hue' parameter to show gender difference
plt.figure(figsize=(10, 6))
sns.scatterplot(x="age", y="salary", hue="gender", data=df, s=150, palette="Set1")
plt.title('Scatter Plot: Salary vs Age (Color by Gender)', fontsize=12, fontweight='bold')
plt.xlabel("Age")
plt.ylabel("Salary ($)")
plt.legend(title="Gender")
plt.tight_layout()
plt.show()
```

## 🎨 Color Palettes in Seaborn

Try these different palettes in your plots:
- `Set1`, `Set2`, `Set3` - Qualitative palettes
- `Blues`, `Greens`, `Oranges` - Sequential palettes
- `coolwarm`, `RdYlBu` - Diverging palettes
- `husl`, `viridis` - Perceptually uniform palettes

**Example:** `sns.scatterplot(..., palette="husl")`
