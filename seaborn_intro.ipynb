## Seaborn vs Matplotlib

|---------|-----------|---------|
| **Level** | Low-level | High-level |
| **Code** | More verbose | More concise |
| **Styling** | Manual | Automatic |
| **DataFrames** | Limited | Native support |

Let's explore Seaborn with some examples!

```python
# Import Required Libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Set the theme/style for all plots
sns.set_theme(style="darkgrid")
print("Theme set to 'darkgrid'")
```

```python
# Create sample data
x = np.array([1, 2, 6, 2, 5, 2, 8, 5, 8, 92, 44])
print(f"X values: {x}")
```

```python
# Calculate y values using sine function
y = np.sin(x)
print(f"Y values (sin): {np.round(y, 3)}")
```

```python
# Create a simple line plot
plt.figure(figsize=(10, 6))
sns.lineplot(x=x, y=y, marker='o', linewidth=2, markersize=8)
plt.title("Simple Line Plot Example", fontsize=14, fontweight='bold')
plt.xlabel("X Values")
plt.ylabel("Sin(X) Values")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Try different styles
available_styles = sns.axes_style().keys()
print(f"Available theme styles: {str(list(sns.axes_style().keys()))}")

# You can try: 'darkgrid', 'whitegrid', 'dark', 'white', 'ticks'
```
